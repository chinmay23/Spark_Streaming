## Task 11. Calculate the count of rows where the gt column contains ‘stairs’ only for users a and b on Device=’nexus4_1’

a. Write and run a streaming query called stairsonlyDF where the output contains a table with the
streaming counts of the number of times the column named gt contains ‘stairs’ anywhere in the
column value, only for users ‘a’ and ‘b’, when they use ‘nexus4_1’ as their Device.

```
stairsonlyDF=streamingDF.where((F.col('Device')== 'nexus4_1') & ((F.col('User')== 'a') | (F.col('User')== 'b')) & 
                               F.col('gt').contains("stairs")).groupby('User').count()
```

```
mystreamingQuery = stairsonlyDF.writeStream.queryName('stairsonly').format("memory").outputMode("complete").start()
```

b. Display /show the streaming output 5 times. The output will have 2 columns – User and count

```
for x in range(5):
  spark.sql("SELECT * FROM stairsonly").show()
  sleep(5)
```

```
mystreamingQuery.stop()
```

##Task 12. Compute the average of colummns x, y, x for each combination of values

called x_avg, the average of y in a column called y_avg, the average of z in a column called

##Task 13. Perform a join of a static DataFrame with a streaming DataFrame

a. We will first group by columns gt and Model, and compute the average values of x, y, z for the
static DataFrame, staticDF, and save the results as historicalAvg. Label the columns as x_avg1,
y_avg1, z_avg1

```
historicalAvg = staticDF.groupby('gt','Model').agg(F.avg("x").alias("x_avg1"),F.avg("y").alias("y_avg1"),F.avg("z").alias("z_avg1"))
```

b. Do the same for the streaming DataFrame streamingDF, and apply a join transformation to join
the historicalAvg DataFrame on columns gt and Model, and write out to memory sink in
complete output mode. Label the columns as x_avg2, y_avg2, z_avg2

```
StreamQuery = streamingDF.groupby('gt','Model').agg(F.avg("x").alias("x_avg2"),F.avg("y").alias("y_avg2"),F.avg("z").alias("z_avg2")).join(historicalAvg, on =['gt','Model'])
```

```
mystreamingQuery = StreamQuery.writeStream.queryName('StrmQuery').format("memory").outputMode("complete").start()
```

c. Display /show the streaming output 5 times. The output will have 8 columns – gt, Model, x_avg1,
y_avg1, z_avg1, x_avg2, y_avg2, z_avg2.

```
for x in range(3):
  spark.sql("SELECT * FROM StrmQuery").show()
  sleep(30)
mystreamingQuery.stop()
```
